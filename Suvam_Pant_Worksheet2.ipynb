Titanic-Database

```python
import pandas as pd
titanic = pd.read_csv("/content/drive/MyDrive/AI/Titanic-Dataset.csv")
```

```python
df = titanic[['Name','Pclass','Sex','Age','Fare','Survived']]
df = df[df['Pclass'] == 1]

print("Mean:", df['Fare'].mean())
print("Median:", df['Fare'].median())
print("Max:", df['Fare'].max())
print("Min:", df['Fare'].min())
```

```python
print("Null Ages:", df['Age'].isnull().sum())

df = df.dropna(subset=['Age'])

```

```python
dummies = pd.get_dummies(titanic['Embarked'], prefix='Embarked')

titanic = pd.concat([titanic, dummies], axis=1)

titanic = titanic.drop(columns=['Embarked'])

titanic.head()
```

```python
import matplotlib.pyplot as plt

titanic.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival by Gender")
plt.ylabel("Mean Survival Rate")
plt.show()
```
