Problem - 3: Array Operations:

```python
#3

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

print("Add:\n", x+y)
print("\nSubtract:\n", x-y)
print("\nMultiply by 3:\n", x*3)
print("\nSquare:\n", x**2)

print("\nDot v·w =", np.dot(v,w))
print("Dot x·v =", np.dot(x,v))
print("Dot x·y =\n", np.dot(x,y))

print("\nConcat x & y (rows):\n", np.concatenate((x,y),axis=0))
print("\nConcat v & w (cols):\n", np.vstack((v,w)))

# Error explanation
try:
    print(np.concatenate((x,v)))
except:
    print("\nERROR: x is 2D and v is 1D so shapes do not match.")
```

Problem - 4: Matrix Operations:

```python
#1
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

A_inv = np.linalg.inv(A)
result = A.dot(A_inv)
print(result)
```

```python
#2
AB = A.dot(B)
BA = B.dot(A)

print("AB =\n", AB)
print("BA =\n", BA)
```

```python
#3
left_side = (A.dot(B)).T
right_side = B.T.dot(A.T)

print("Left Side (AB)^T =\n", left_side)
print("Right Side B^T A^T =\n", right_side)
```

```python
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A)
X = A_inv.dot(B)
print("Solution:", X)
```

Experiment: How Fast is Numpy?

```python
import time

# 1. ELEMENT-WISE ADDITION
# Python lists
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))

start = time.time()
list_add = [list1[i] + list2[i] for i in range(1_000_000)]
end = time.time()

print("Python List Addition Time:", end - start, "seconds")


# NumPy arrays
arr1 = np.arange(1_000_000)
arr2 = np.arange(1_000_000)

start = time.time()
arr_add = arr1 + arr2
end = time.time()

print("NumPy Addition Time:", end - start, "seconds")


# 2. ELEMENT-WISE MULTIPLICATION
# Python lists
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(1_000_000)]
end = time.time()

print("Python List Multiplication Time:", end - start, "seconds")


# NumPy
start = time.time()
arr_mul = arr1 * arr2
end = time.time()

print("NumPy Multiplication Time:", end - start, "seconds")



# 3. DOT PRODUCT
# Python list dot product
start = time.time()
list_dot = sum(list1[i] * list2[i] for i in range(1_000_000))
end = time.time()

print("Python List Dot Product Time:", end - start, "seconds")


# NumPy dot product
start = time.time()
arr_dot = np.dot(arr1, arr2)
end = time.time()

print("NumPy Dot Product Time:", end - start, "seconds")


# 4. MATRIX MULTIPLICATION
# Python lists
mat1 = [[i for i in range(1000)] for _ in range(1000)]
mat2 = [[i for i in range(1000)] for _ in range(1000)]

start = time.time()

# Manual matrix multiplication (slow)
result = [[sum(mat1[i][k] * mat2[k][j] for k in range(1000))
           for j in range(1000)] for i in range(1000)]

end = time.time()
print("Python List Matrix Multiplication Time:", end - start, "seconds")


# NumPy matrix multiplication
A = np.random.rand(1000, 1000)
B = np.random.rand(1000, 1000)

start = time.time()
C = np.dot(A, B)
end = time.time()

print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
